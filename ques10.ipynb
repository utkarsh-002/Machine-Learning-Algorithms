```python
import pandas as pd
df = pd.read_csv('./Datasets/ColourXY.csv')
df.head()
```

```python
df=df.drop(['color'],axis=1)
df.head()
```

```python
df.info(verbose=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test = train_test_split(df,test_size=.3)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcc_list=[]

for i in range(1,11):
    model = KMeans(n_clusters=i)
    model.fit(x_train)
    wcc_list.append(model.inertia_)
    
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcc_list)
plt.title('elbow method graph')
plt.xlabel('Number of cluster')
plt.ylabel('WCC Score')
```

```python
x = df[['x','y']]
```

```python
model =   KMeans(n_clusters=3)
model.fit(x)
pred = model.predict(x_test)
pred
```

```python
import seaborn as sns
sns.scatterplot(x='x',y='y',data=x,hue=model.labels_)
```
