```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
data = pd.read_csv("ColourXY.csv")
data.head()
```

```python
from sklearn.model_selection import train_test_split
x = data[['x','y']]
y = data['color']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
kl = KMeans()
kl.fit(x_train,y_train)
new_data = [[155,176]]
print(kl.predict(new_data))
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['x'],data['y'],c=data['color'],cmap='rainbow')
```

```python
wcss = []
for i in range(1,11):
  k = KMeans(n_clusters=i,init="k-means++",random_state=42)
  k.fit(x)
  wcss.append(k.inertia_)
plt.plot(range(1,11),wcss)
plt.show()  
```

```python
k = KMeans(n_clusters=3)
plt.scatter(data['x'],data['y'],c=k.fit_predict(x),cmap='rainbow')
plt.show()
```
