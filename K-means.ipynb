### Importing the necessasry packages and libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

### Reading the dataset 

```python
data = pd.read_csv('./datasets/basic4.csv')
print(data)
print("\nUnique values: ", data['color'].unique())
```

### Processing the data such that its a tuple of 'x' and 'y' for each input

```python
X = tuple(zip(data['x'], data['y']))
X
```

### Plotting the dataset for reference

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Dataset')
plt.scatter(data['x'], data['y'],c = data['color'], cmap ='rainbow')
plt.show()
```

### Using the WCSS method to find the k-value [WCSS: Within-Cluster Sum of Squares]

```python
wcss = []
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(X)  
    wcss.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss)  
plt.title('The Elbow Method Graph')  
plt.xlabel('Number of clusters(k)')  
plt.ylabel('wcss_list')  
plt.show() 
```

### As seen from the graph above the value  is 3, creating a model with 3 clusters

```python
kmeans = KMeans(n_clusters=3)
plt.title('Clusters=3')
plt.xlabel('X')
plt.ylabel('Y')
plt.scatter(data['x'], data['y'],c = kmeans.fit_predict(X), cmap ='rainbow')
plt.show()
```

# Inference
- The dataset contains points that belong to 3 groups 
- First, pre-process the data so that we get the x and y values for each row/input
- Then use Within-Cluster Sum of Squares(WCSS) to find the point (Elbow Method)
- Since 3 is the point where we find the 'elbow' , 3 is the ideal k-value
- Make a model with 3 clusters 
- Fitting and predicting it with the input 
- The graph shows that it clearly distinguished the datapoints into 3 different clusters and the prediction is very close to the original groups
