```python
import pandas as pd
df = pd.read_csv('./Datasets/diabetes3.csv')
df.head()
```

```python
df.isna().sum()
```

```python
import seaborn as sns
sns.pairplot(df,x_vars=['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age'],y_vars=['Outcome'])
```

```python
import matplotlib.pyplot as plt
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(9,9))
sns.heatmap(df[top_corr_features].corr(),annot=True)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
df['Glucose']=sc.fit_transform(df['Glucose'].values.reshape(-1,1))
df['BMI']=sc.fit_transform(df['BMI'].values.reshape(-1,1))
```

```python
df['Age']=sc.fit_transform(df['Age'].values.reshape(-1,1))
df=df.drop(['BloodPressure','SkinThickness'],axis=1)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop(['Outcome'],axis=1),df['Outcome'])
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
xtest=[[20,0.78,0,0.56,.56,0]]
pre = model.predict(xtest)
print("person has diabetes"if pre else "Person don't have diabetes")
```

```python
from sklearn.metrics import confusion_matrix,roc_curve,classification_report
pred = model.predict(x_test)
confusion_matrix(y_test,pred)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(model,x_test,y_test)
```

```python
print(classification_report(y_test,pred))
```

```python
fpr={}
tpr={}
thresh={}
proba = model.predict_proba(x_test)
for i in range(2):
    fpr[i],tpr[i],thresh[i]=roc_curve(y_test,proba[:,i],pos_label=i)
plt.plot(fpr[0],tpr[0],linestyle='-',c='green')
plt.plot(fpr[1],tpr[1],linestyle='-',c='blue')
plt.plot([[0,0],[1,1]],linestyle='dashed',c='orange')
plt.show
```
