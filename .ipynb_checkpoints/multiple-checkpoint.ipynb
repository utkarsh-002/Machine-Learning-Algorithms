```python
import pandas as pd
import numpy as np
data=pd.read_csv('advertising.csv')
data.head()
```

```python
data=data.drop(['Unnamed: 0'],axis=1)
```

```python
x=data.drop(['Sales'],axis=1)
y=data['Sales'].values.reshape(-1,1)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
model.fit(x_train,y_train)
```

```python
print("Slope: ",model.coef_)
print("Intercept: ",model.intercept_)
```

```python
print("the linear model is y= {:.5}+{:.5}*TV+{:.5}*Radio+{:.5}*Newspaper".format(model.intercept_[0],model.coef_[0][0],model.coef_[0][1],model.coef_[0][2]))
```

```python
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
%matplotlib inline
sns.pairplot(data, x_vars=['TV','Radio','Newspaper'],y_vars=['Sales'],kind='reg',size=7)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
pred=model.predict(x_test)
mae=mean_absolute_error(y_test,pred)
mae
```

```python
mse=mean_squared_error(y_test,pred)
mse
```

```python
rmse=np.sqrt(mean_squared_error(y_test,pred))
rmse
```
